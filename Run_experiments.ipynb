#### **Run experiments on toy datasets**

```
# # Downloading model weights
%%shell
eval "$(conda shell.bash hook)" # copy conda command to shell
conda activate character-bert

python preprocessing.py --data "atis"

conda deactivate
```

```
# # Downloading model weights
%%shell
eval "$(conda shell.bash hook)" # copy conda command to shell
conda activate character-bert

python download.py --model='general_bert'
python download.py --model='general_character_bert'
python download.py --model='bert-base-uncased'
#python download.py --model='medical_character_bert'

conda deactivate
```

```
%%shell
eval "$(conda shell.bash hook)" # copy conda command to shell
conda activate character-bert

# NER with BERT
python main.py \
    --task='sequence_labelling' \
    --embedding='general_character_bert' \
    --train_batch_size 10 \
    --eval_batch_size 10 \
    --num_train_epochs 40 \
    --validation_ratio 0.2 \
    --train_size 20000 \
    --test_size 10000 \
    --do_lower_case \
    --do_train \
    --do_predict
conda deactivate
```
